````python
import pandas as pd
import numpy as np

def get_energy():
    df1 = pd.read_excel('Energy Indicators.xls')    
    #rename some columns
    df1 = df1.rename(columns={'Unnamed: 1': 'Country','Unnamed: 3': 'Energy Supply','Unnamed: 4': 'Energy Supply per Capita','Unnamed: 5':'% Renewable'})
    #use these columns only
    cols = ['Country','Energy Supply','Energy Supply per Capita','% Renewable']
    df1 = df1[cols]
    #drop the NAN
    df1 = df1.dropna()
    #replace "..." with np.NAN
    df1 = df1.replace('...',np.nan)
    #convert 'Energy supply' in Giga-Joules 1000000 Giga Joule = Peta Joule
    df1['Energy Supply'] = 1000000 * df1['Energy Supply']
    #rename some countries
    dict = {"Republic of Korea": "South Korea",
        "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"}
    #,
    #    'Bolivia (Plurinational State of)':'Bolivia'}
    df1 = df1.replace({'Country': dict})
    #remove everything that is in (), including the parenteses
    df1 = df1.replace(r' \([^)]*\)', '',regex=True)
    #set index
    #df1 = df1.set_index('Country')

    #print(df1.columns)
    #print(df1.loc['Portugal'])
    #print(df1.dtypes)
    return df1


def get_GDP():
    df2 = pd.read_csv('world_bank.csv')
    #rename column related to countries    
    df2 = df2.rename(columns={'Data Source': 'Country'})
    #rename some countries
    dict_GDP = {"Korea, Rep.": "South Korea", "Iran, Islamic Rep.": "Iran","Hong Kong SAR, China": "Hong Kong"}
    df2 = df2.replace({'Country': dict_GDP})
    #df2 = df2.set_index('Country')
    return df2

def get_ScimEn():
    df3 = pd.read_excel('scimagojr-3.xlsx')
    return df3

def answer_one():
    energy = get_energy()
    GDP = get_GDP()
    ScimEn = get_ScimEn()
    
    #conditions:
    #--- Use only the last 10 years (2006-2015) of GDP data
    
    GDP = GDP.rename(columns={'Unnamed: 50':'2006','Unnamed: 51':'2007','Unnamed: 52':'2008',
                    'Unnamed: 53':'2009','Unnamed: 54':'2010','Unnamed: 55':'2011','Unnamed: 56':'2012',
                    'Unnamed: 57':'2013','Unnamed: 58':'2014','Unnamed: 59':'2015'})
    
    cols_GDP = ['Country','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']
    GDP = GDP[cols_GDP]
    #--- only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15)
    ScimEn = ScimEn.head(15)
    ScinEn = ScimEn.reset_index()
    ScimEn = ScimEn.set_index('Country')
    
    df_m1 = pd.merge(energy,GDP,how='inner',left_on='Country', right_on='Country')
    
    df =pd.merge(ScimEn,df_m1,how='inner',left_index=True,right_on='Country')
    
    df = df.set_index('Country')
    #['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    
    return df

def answer_twelve():
    Top15 = answer_one()

    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
        
    Top15['Country'] = Top15.index
    
    Top15['Continent'] = Top15['Country'].map(ContinentDict)
    
    Top15['bins'] = pd.cut(Top15['% Renewable'],5)
    
    Top15 = Top15.set_index('Continent','% Renewable')
    
    Top15 = Top15.groupby([Top15.index,'bins']).size()
    
    return Top15
answer_twelve()
````
